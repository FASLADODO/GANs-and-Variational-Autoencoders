# Cost Function (part 2)<hr>
- Normally, we would have gone straight to implementation by now, but there are more important points to cover
- Not obvious: there are some problems with the cost function we just defined

## Problem
- The cost function from perspective of the generator is problematic
![problem_cost_function](../images/problem_cost_function.PNG)
The first term corresponds to real images and the second term corresponds to fake images generated by the generator.<br>
The first term with respect to the generator phram is zero because the generator phram don't appear anywhere in that term.<br>
So the only relevant part from the perspective of the generator is the second term.


- Gradient of 1st term wrt \\(\theta_G\\) is 0
- So only the 2nd term is relevant

#### Problem (cont.)
- Suppose the discriminator is very successful at telling the difference between real and fake
- Then D(G(z)) is very close to 0 
- The slope at this point also very close to 0
- Problematic because \\(\theta_G\\) changes in a direction proportional to gradient
- When the discriminator is too good, the generator has no chance to improve ! 
![cost_problem_2](../images/cost_problem_2.PNG)

<hr>
## Solution
- Use a different cost function for the generator
- " Flip the target "


- Instead of a target 0 for fake images, the generator wants the target to be 1
- Makes sense because the term log(D(any image)) corresponds to the log-likelihood of the target being 1 

![cost_problem_3](../images/cost_problem_3.PNG)

#### Notes
- Since \\(J^{(G)} + J^{(D)} != 0 \\), this is no longer a zero-sum game (any theoretical analysis done on such a game no longer applies)
- You may see this "fixed" generator cost referred to as a "non-saturating heuristic"
- "Non-saturating" -> gradient doesn't fall to 0
- "Heuristic" -> no theoretical (probabilistic) reasoning, just solves a numerical issue
